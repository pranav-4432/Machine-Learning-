```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
x=[
    [180,7],
    [200,7.5],
    [250,8],
    [300,8.5],
    [330,9],
    [360,9.5]  
]

y=[0,0,0,1,1,1]  #0= apple, 1=Orange

model=KNeighborsClassifier(n_neighbors=3)  #analyzing around 3 points

model.fit(x,y)
weight=float(input("Enter the weight in grams"))
size=float(input("Enter the size in cm"))

prediction=model.predict([[weight,size]])[0]

if prediction==0:
    print("This is an likely Apple")
else:
    print("This is likely an orange")
```

# metrics

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
#true answers(What actually happened)
y_true=[1,0,1,1,0,1,0]
```

```python
#model prediction (what it guessed)
y_pred=[1,0,1,0,0,1,1]
```

```python
#evaluation
print("Accuracy: ",accuracy_score(y_true,y_pred))
print("Precision: ",precision_score(y_true,y_pred))
print("Recall: ",recall_score(y_true,y_pred))
print("F1 score: ", f1_score(y_true,y_pred))
```

# MAE,MSE,RMSE

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
import numpy as np
```

```python
#real scores
real_scores=[90,60,80,100]
```

```python
#model guess
predicted_scores=[85,70,70,95]
```

```python
mae= mean_absolute_error(real_scores,predicted_scores)
mse= mean_squared_error(real_scores,predicted_scores)
rmse=np.sqrt(mse)
```

```python
print("MAE: On average off by:",mae)
print("MSE: On average off by:",mse)
print("RMSE: On average off by:",rmse)
```

```python
def reverse_number(num):
    reverse=0

    while num>0:
        digit = num%10
        reverse=reverse*10+digit
        num=num// 10
    return reverse
```

```python
result=reverse_number(153)
```
